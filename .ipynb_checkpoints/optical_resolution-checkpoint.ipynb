```python
import pandas as pd
import math
```

```python
calib = pd.read_csv("get3alpha.csv", usecols=[0,3], names=["qcal","qped"], encoding="utf8")
sigma = pd.read_csv("sigmas.csv", usecols=[0], names=["sigma"], encoding="utf8")
```

```python
ch_sum = list(range(1,64+1,1))
dssd_ch=[i for i in ch_sum]
Eoptical_resolution=[calib["qcal"]*(sigma["sigma"]-calib["qped"])*1000*2.35]

custom_style = [
    {
        'selector': 'th',
        'props':[
            ('border', 'solid black 1px'), 
            ('text-align', 'center'), 
        ]
    },
    {
        'selector': 'td', 
        'props':[
            ('border', 'solid black 1px'), 
            ('text-align', 'center'), 
        ]
    }
]

df = pd.DataFrame(Eoptical_resolution)
name = ['エネルギー分解能']
df.index=name
df.columns =dssd_ch
df = df.T
df.style.set_table_styles(custom_style)
```
